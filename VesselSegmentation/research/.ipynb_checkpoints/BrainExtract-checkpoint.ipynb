```python
def get_distanse_matrix(R: int)->torch.tensor:
    dist_matrix = torch.zeros(3*(2*R+1,))
    for i in range(-R, R+1):
        for j in range(-R, R+1):
            for k in range(-R, R+1):
                dist_matrix[i+R,j+R,k+R]=i*i+j*j+k*k
    dist_matrix = dist_matrix**0.5
    return(dist_matrix)

def pad_img(img:torch.tensor, R: int)->torch.tensor:
    s = torch.tensor(img.shape)
    new_img = torch.zeros(s[0]+2*R, s[1]+2*R, s[2]+2*R)
    new_img[R:R+s[0], R:R+s[1], R:R+s[2]] = img
    return(new_img)

def unpad_img(img:torch.tensor, R: int)->torch.tensor:
    s = torch.tensor(test_img.shape) - 2*R
    return(img[R:R+s[0], R:R+s[1], R:R+s[2]])

def get_dist(img_part, dist_matrix, R):
    tmp = img_part*dist_matrix
    if len(tmp[torch.where(tmp>0)]):
        return min(tmp[torch.where(tmp>0)])
    else:
        return 3**0.5*R
```

```python
GT_img = vessels_img#[200:300, 300:400]
SEG_img = nn_img#[200:300, 300:400]
```

```python
def get_dist_img(img, R):
    t = time.time()
    SIZE = img.shape
    dist_matrix = get_distanse_matrix(R)
    img_pad = pad_img(img, R)
    dist_img = torch.zeros(SIZE)
    for i in range(SIZE[0]):
        for j in range(SIZE[1]):
            for k in range(SIZE[2]):
                if img_pad[i+R, j+R, k+R]==0:
                    dist_img[i, j, k] = get_dist(img_pad[i:i+2*R+1, j:j+2*R+1, k:k+2*R+1],
                                           dist_matrix, R)
    print(time.time() - t)
    return(dist_img)
```
